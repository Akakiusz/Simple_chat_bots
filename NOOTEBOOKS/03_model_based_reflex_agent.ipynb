### Model Based Reflex Agent 

A model based reflex agent is an upgrade from a simple reflex agent. 
The key difference is that it keeps track of the current state of the world — it has **memory**. 
This means it can make decisions based not just on the current input, but also on what happened before.



```python
#   

class ModelBasedReflexAgent:   

    def __init__(self, darkness_threshold=400, memory_length=5):
        self.darkness_threshold = darkness_threshold
        self.memory_length = memory_length
        self.memory = []

    def perceive(self, light_level):
        self.memory.append(light_level)
        if len(self.memory) > self.memory_length:
            self.memory.pop(0)

```
