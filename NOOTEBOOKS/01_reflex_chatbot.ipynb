## Project Description

This project automates a simple customer support process for a logistics service similar to DHL.

Customers frequently ask about shipment tracking, delivery times, and delays. 

## User Guidance

The chatbot isdeddicated to provide support for delivery-related queries. 
Users should interact with the system using simple and clear phrases.

#### Supported Queries
The chatbot can handle the following types of requests:
- Tracking a shipment
- Asking about delivery times
- Reporting delays
- Requesting general help

#### Example Commands
Users can interact with the chatbot using commands such as:
- "Track my shipment"
- "Where is my package?"
- "Is my delivery delayed?"
- "Help"

#### Usage Guidelines
- Use clear and simple language
- Avoid complex and long sentences
- If required please provide necessary details

#### Limitations
Chatbot cannot:
- Understand complex or ambiguous queries
- Provide personalised or real-time tracking data
- Learn from previous interactions

#### Finish Chat
- Please use phrase "bye"


```python
# Importing necessary libraries

import random
```

### Import library

The *`random`* library is imported from Python libraries to generate random responses. 

```python
# Predefined responses for the chatbot

chatbot_greetings = [ 
    "Hey! Welcome! I promise your package is fine. Mostly",
    "Hi there! What can I do quickly for you?",
    "Hello Dear User! Nice to meet you! How may I help?",
    "Hello! I am DHL bot. I know where your package is. Probably.",
    "Hey! Need help with your shipment?",
    "Greetings! How can I support you today?",
    "Hi! Ask me about your package.",
]
```

```python
# Responses for tracking inquiries

chatbot_tracking_responses = [
    "Your package is currently in transit.",
    "Your shipment is on the way.",
    "Tracking shows your package is moving.",
    "Your parcel has left the warehouse. It is on an adventure.", 
    "Good news! Your package is in transit and heading your way.", 
    "Your package is traveling. It is basically on holiday right now."
]
```

```python
# Responses for timeframe inquiries

chatbot_timeframe_responses = [
    "Delivery usually takes 3-5 business days.",
    "Your package should arrive within a few days.",
    "Shipping time is typically between 3 and 5 days.", 
    "3 to 5 business days. Please do not refresh the tracking page every 10 minutes.",
    "Your package should arrive in 3-5 days. Go touch some grass while you wait."
]
```

```python
# Responses for delayed shipments

chatbot_delayed_responses = [
    "Your shipment may be delayed due to high demand.",
    "We apologise, your delivery is delayed.",
    "There is a slight delay in your shipment.",
    "Delay confirmed. We are very sorry. Your package is not.",
    "Unfortunately there is a delay. Your parcel is thinking about its life choices."
]
```

```python
# Responses for lost or canceled shipments

chatbot_canceled_lost_responses = [
    "Your shipment has been canceled. Please contact support for more information.",
    "We regret to inform you that your package has been lost. Please reach out to our customer service for assistance.",
    "Unfortunately, your delivery has been canceled. Please contact us for further details."
]       
```

```python
# Responses for help inquiries

help_responses = [
    "You can ask about tracking, delays, or delivery.",
    "Try asking: 'Where is my package?' or 'Is my delivery delayed?'",
    "I can help you track your shipment or check delivery time."
]
```

```python
# Responses for out-of-scope inquiries

chatbot_out_of_scope_responses = [
    "Sorry, I dont understand. Please try again.",
    "I am not sure what you mean. Can you rephrase?",
    "Please ask about delivery or tracking.",
    "Please ask about your package or delivery status."
]

print("DHL Assistant: I am your DHL assistant. Ask me about your package.")
```

### Defined user input and responses

```python
while True:
    user_input = input("You: ")
   
    if user_input.lower() == "hello" or user_input.lower() == "hi" or user_input.lower() == "hey":
        print("DHL Assistant:", random.choice(chatbot_greetings))
   
    elif user_input.lower() == "how are you":
        print("DHL Assistant: I am just a DHL assistant, but I am doing pretty well. Thank you!")

    elif "track" in user_input.lower():
        print("DHL Assistant:", random.choice(chatbot_tracking_responses))
   
    elif "delay" in user_input.lower() or "delayed" in user_input.lower() or "late" in user_input.lower():
        print("DHL Assistant:", random.choice(chatbot_delayed_responses))

    elif "timeframe" in user_input.lower() or "delivery time" in user_input.lower():
        print("DHL Assistant:", random.choice(chatbot_timeframe_responses))
    
    elif "lost" in user_input.lower() or "canceled" in user_input.lower():
        print("DHL Assistant:", random.choice(chatbot_canceled_lost_responses)) 

    elif "help" in user_input.lower():
        print("DHL Assistant:", random.choice(help_responses))

    elif user_input.lower() == "bye":
        print("DHL Assistant: Goodbye! Have a great day!")
        break
   
    else:
        print("DHL Assistant:", random.choice(chatbot_out_of_scope_responses))
```

### This chatbot is a simple reflex agent. It responds to user input based on predefined rules. No using memory or learning.
