## Project Description

This project automates a simple customer support process for an online store.

## User Guidance

The chatbot isdeddicated to provide support for delivery-related queries. 
Users should interact with the system using simple and clear phrases.

#### Supported Queries
The chatbot can handle the following types of requests:
- Tracking a shipment
- Asking about delivery times
- Reporting delays
- Requesting general help

#### Example Commands
Users can interact with the chatbot using commands such as:
- "Track my shipment"
- "Where is my package?"
- "Is my delivery delayed?"
- "Help"

#### Usage Guidelines
- Use clear and simple language
- Avoid complex and long sentences
- If required please provide necessary details

#### Limitations
Chatbot cannot:
- Understand complex or ambiguous queries
- Provide personalised or real-time tracking data
- Learn from previous interactions


```python
import random
```

### Import library

The *`random`* library is imported from Python libraries to generate random responses. 

```python
greetings = ["Hello!", "Hi there!", "Hey!", "Hi!"]

print("Chatbot: Hi! I am a simple chatbot. How can I help you today?")

```

#

```python
while True:
    user_input = input("You: ")
   
    if user_input.lower() == "hello":
        print("Chatbot:", random.choice(greetings))
   
    elif user_input.lower() == "how are you?":
        print("Chatbot: I am just a program, but I’m doing well. Thank you!")
   
    elif user_input.lower() == "bye":
        print("Chatbot: Goodbye! Have a great day!")
        break
   
    else:
        print("Chatbot: Sorry, I don’t understand that.")
```
