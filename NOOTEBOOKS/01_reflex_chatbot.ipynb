## Project Description

This project automates a simple customer support process for a logistics service similar to DHL.

Customers frequently ask about shipment tracking, delivery times, and delays. 

## User Guidance

The chatbot isdeddicated to provide support for delivery-related queries. 
Users should interact with the system using simple and clear phrases.

#### Supported Queries
The chatbot can handle the following types of requests:
- Tracking a shipment
- Asking about delivery times
- Reporting delays
- Requesting general help

#### Example Commands
Users can interact with the chatbot using commands such as:
- "Track my shipment"
- "Where is my package?"
- "Is my delivery delayed?"
- "Help"

#### Usage Guidelines
- Use clear and simple language
- Avoid complex and long sentences
- If required please provide necessary details

#### Limitations
Chatbot cannot:
- Understand complex or ambiguous queries
- Provide personalised or real-time tracking data
- Learn from previous interactions


```python
import random
```

### Import library

The *`random`* library is imported from Python libraries to generate random responses. 

```python
chatbot_greetings = [ 
    "Hello! How can I assist you today?",
    "Hi there! What can I do for you?",
    "Hey! Nice to meet you! How may I help?"
    "Hi! How can I help you with your delivery?",
    "Hey! Need help with your shipment?",
    "Greetings! How can I support you today?",
    "Hi! Ask me about your package.",
]
```

```python
chatbot_tracking_responses = [
    "Your package is currently in transit.",
    "Your shipment is on the way.",
    "Tracking shows your package is moving.",
    "Your parcel has left the warehouse."]
```

```python
chatbot_out_of_scope_responses = [
    "Sorry, I dont understand. Please try again.",
    "I am not sure what you mean. Can you rephrase?",
    "Please ask about delivery or tracking."
]

print("DHL Assistant: I am your DHL assistant. Ask me about your package.")
```

### Defined user input and responses

```python
while True:
    user_input = input("You: ")
   
    if user_input.lower() == "hello":
        print("DHL Assistant:", random.choice(chatbot_greetings))
   
    elif user_input.lower() == "how are you?":
        print("DHL Assistant: I am just a program, but I’m doing well. Thank you!")
   
    elif user_input.lower() == "bye":
        print("DHL Assistant: Goodbye! Have a great day!")
        break
   
    else:
        print("DHL Assistant: Sorry, I don’t understand that.")
```
