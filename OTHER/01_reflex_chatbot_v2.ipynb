## Project Description

This project automates a simple customer support process for a logistics service similar to DHL.

Customers frequently ask about shipment tracking, delivery times, and delays.

## User Guidance

The chatbot is dedicated to provide support for delivery-related queries.
Users should interact with the system using simple and clear phrases.

#### Supported Queries
- Tracking a shipment
- Asking about delivery times
- Reporting delays
- Lost or canceled packages
- Requesting general help

#### Limitations
Chatbot cannot:
- Understand complex or ambiguous queries
- Provide real-time tracking data
- Learn from previous interactions

### Response function

The `get_response()` function handles all the logic.

For tracking queries, it first asks for a tracking number, then on the next message gives the tracking status. A simple variable `waiting_for_tracking_number` is used to remember what step we are on.

```python
# This variable remembers if we are waiting for a tracking number
waiting_for_tracking_number = False


def get_response(user_input):
    global waiting_for_tracking_number
    text = user_input.lower()

    # If we already asked for a tracking number, treat next input as the number
    if waiting_for_tracking_number:
        waiting_for_tracking_number = False
        return random.choice(chatbot_tracking_responses)

    if text in ("hello", "hi", "hey"):
        return random.choice(chatbot_greetings)

    elif "how are you" in text:
        return random.choice(chatbot_how_are_you_responses)

    elif "track" in text or "where is my" in text or "where is my package" in text:
        waiting_for_tracking_number = True
        return random.choice(chatbot_ask_tracking_number)

    elif "delay" in text or "late" in text:
        return random.choice(chatbot_delayed_responses)

    elif "timeframe" in text or "delivery time" in text or "how long" in text:
        return random.choice(chatbot_timeframe_responses)

    elif "lost" in text or "canceled" in text or "cancel" in text:
        return random.choice(chatbot_canceled_lost_responses)

    elif "help" in text:
        return random.choice(help_responses)

    else:
        return random.choice(chatbot_out_of_scope_responses)
```

```python
while True:
    user_input = input("You: ")

    if user_input.lower() == "bye":
        print("DHL Assistant: Goodbye! Have a great day!")
        break

    response = get_response(user_input)
    print("DHL Assistant:", response)
```
